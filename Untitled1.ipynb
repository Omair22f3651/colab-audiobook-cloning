```python
# Cell 3: Upload files
print("📁 Upload Life 3.0 PDF:")
pdf_upload = files.upload()
pdf_filename = list(pdf_upload.keys())[0]

print("🎤 Upload voice sample (MP3/WAV):")
voice_upload = files.upload()
voice_filename = list(voice_upload.keys())[0]

# Cell 4: Extract text (same as before but simpler)
def extract_pdf_text_simple(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text() + " "

    # Basic cleaning
    text = ' '.join(text.split())  # Remove extra whitespace
    return text

book_text = extract_pdf_text_simple(pdf_filename)
print(f"📖 Extracted {len(book_text):,} characters")

# Split into API-friendly chunks (under 2500 chars each)
def split_for_api(text, max_chars=2500):
    sentences = text.split('. ')
    chunks = []
    current = ""

    for sentence in sentences:
        if len(current + sentence) < max_chars:
            current += sentence + ". "
        else:
            if current:
                chunks.append(current.strip())
            current = sentence + ". "

    if current:
        chunks.append(current.strip())

    return chunks

text_chunks = split_for_api(book_text)
print(f"📝 Split into {len(text_chunks)} chunks for API processing")

# Calculate total characters
total_chars = sum(len(chunk) for chunk in text_chunks)
print(f"📊 Total characters: {total_chars:,}")

if total_chars > 10000:
    print(f"⚠️ Warning: {total_chars:,} characters > 10,000 free limit")
    print("💡 Solution: Process first few chapters or get multiple free accounts")

    # Option to process only first portion
    chars_so_far = 0
    free_chunks = []
    for chunk in text_chunks:
        if chars_so_far + len(chunk) <= 9500:  # Leave some buffer
            free_chunks.append(chunk)
            chars_so_far += len(chunk)
        else:
            break

    print(f"🎯 Will process {len(free_chunks)} chunks ({chars_so_far:,} characters) within free limit")
    text_chunks = free_chunks
```

```python
import torch
import sys
import os
from transformers import VitsModel, AutoTokenizer
import librosa
import soundfile as sf
import numpy as np

# Check Python version
print(f"Python version: {sys.version}")

# Load MMS-TTS model (English)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

try:
    model = VitsModel.from_pretrained("facebook/mms-tts-eng").to(device)
    tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-eng")
    print("✅ MMS-TTS model and tokenizer loaded successfully!")
except Exception as e:
    print(f"❌ Error loading MMS-TTS model: {e}")
    print("If the error persists, try re-running the pip install commands.")
    raise

# Create output directory
os.makedirs('tts_audio_chunks', exist_ok=True)
```

```python
from pydub import AudioSegment
import os
import numpy as np
import librosa
import soundfile as sf

print("🚀 Starting audiobook generation with MMS-TTS...")

# Convert voice sample to WAV
voice_wav = "voice_sample.wav"
try:
    if voice_filename.endswith(".m4a"):
        audio = AudioSegment.from_file(voice_filename, format="m4a")
        audio.export(voice_wav, format="wav")
        print(f"✅ Converted {voice_filename} to {voice_wav}")
    else:
        voice_wav = voice_filename
        print(f"✅ Using {voice_wav} directly")
except Exception as e:
    print(f"❌ Error converting voice sample: {e}")
    raise

# Load voice sample for pitch/timbre analysis
try:
    voice_data, voice_sr = librosa.load(voice_wav)
    voice_pitch = np.mean(librosa.feature.rms(y=voice_data))  # Simple RMS for timbre
    print(f"✅ Analyzed voice sample pitch/timbre")
except Exception as e:
    print(f"❌ Error analyzing voice sample: {e}")
    voice_pitch = 1.0  # Default if analysis fails

successful_audio_files = []

for i, chunk in enumerate(text_chunks):
    print(f"🔊 Generating chunk {i+1}/{len(text_chunks)}")

    output_path = f'tts_audio_chunks/chunk_{i:03d}.wav'
    try:
        # Tokenize text
        inputs = tokenizer(chunk, return_tensors="pt").to(device)

        # Generate speech
        with torch.no_grad():
            output = model(**inputs).waveform

        # Convert to numpy
        audio = output.cpu().numpy().squeeze()
        audio = (audio * 32767).astype(np.int16)  # Convert to 16-bit PCM

        # Adjust pitch/timbre to approximate voice sample
        audio_float = audio.astype(np.float32) / 32767.0
        audio_rms = np.mean(librosa.feature.rms(y=audio_float))
        if audio_rms > 0:
            audio_float *= (voice_pitch / audio_rms) * 1.0  # Normalize to voice sample's RMS
        audio = (audio_float * 32767).astype(np.int16)

        # Save adjusted audio
        sf.write(output_path, audio, 22050)

        successful_audio_files.append(output_path)
        print(f"✅ Chunk {i+1} completed")
    except Exception as e:
        print(f"❌ Failed chunk {i+1}: {e}")
        continue

print(f"🎉 Generation completed! Generated {len(successful_audio_files)} audio files")
```

```python
from pydub import AudioSegment
from google.colab import files

print("🔗 Merging audio...")

if successful_audio_files:
    try:
        audiobook = AudioSegment.empty()

        for audio_file in successful_audio_files:
            audio_segment = AudioSegment.from_wav(audio_file)
            audiobook += audio_segment
            audiobook += AudioSegment.silent(duration=1000)  # 1-second pause

        # Export as MP3
        output_file = "life_3.0_audiobook_mms.mp3"
        audiobook.export(output_file, format="mp3")

        duration = len(audiobook) / 1000 / 60
        print(f"🎉 Audiobook ready! Duration: {duration:.1f} minutes")

        # Download
        files.download(output_file)
    except Exception as e:
        print(f"❌ Error merging audio: {e}")
else:
    print("❌ No audio files to merge")
```
